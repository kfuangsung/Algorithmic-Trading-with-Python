```python
from backtester import *
```

```python
# two Moving Average
path = 'forex_prices/m5/EUR_USD_m5.csv'
freq = path.split('.')[0].split('_')[-1]
data = pd.read_csv(path, parse_dates=['date'], index_col=['date'])
periods = (300, 400)
kind = 'EMA'
backtester = TwoMovingAverageBacktester(data=data, freq=freq, periods=periods,kind=kind)
```

```python
backtester.backtest()
```

```python
backtester.results()
```
