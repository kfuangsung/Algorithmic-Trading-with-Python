```python
import sys 
sys.path.append('../')
import pandas as pd 
import numpy as np 
import ta
import sqlite3 as sql
import matplotlib.pyplot as plt 
from backtester import * 
from fxcmtoken import major_forex_pairs
```

```python
major_forex_pairs
```

```python
conn = sql.connect('../PriceData/PriceData_m5.db')
ticker = major_forex_pairs[0]
data = pd.read_sql(f"SELECT * FROM '{ticker}'", conn, parse_dates=['date'], index_col=['date'])
for col in ['open', 'close', 'high', 'low']:
    data[col] = data[['bid'+col, 'ask'+col]].mean(axis=1)
conn.close()
data
```

```python
window = 15
data['trix'] = ta.trend.trix(close=data['close'], window=window)
data.dropna(inplace=True)
```

```python
data['trix'].describe()
```

```python
data[['close', 'trix']].iloc[-200:].plot(figsize=(15,5), secondary_y='trix')
plt.show()
```

```python
data['signal'] = np.where(data['trix'] > 0, 1,
                          np.where(data['trix'] < 0, -1, 0))
```

```python
data['signal'].value_counts()
```

```python
data['signal'].diff(1).apply(np.abs).value_counts()
```

```python
backtest = IterativeBacktester(data, data['signal'], 'm5')
backtest.backtest()
backtest.results()
```
