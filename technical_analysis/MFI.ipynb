```python
import sys 
sys.path.append('../')
import pandas as pd 
import numpy as np 
import ta
import sqlite3 as sql
import matplotlib.pyplot as plt 
from backtester import * 
from fxcmtoken import major_forex_pairs
```

```python
print(major_forex_pairs)
conn = sql.connect('../PriceData/PriceData_m5.db')
ticker = major_forex_pairs[0]
data = pd.read_sql(f"SELECT * FROM '{ticker}'", conn, parse_dates=['date'], index_col=['date'])
for col in ['open', 'close', 'high', 'low']:
    data[col] = data[['bid'+col, 'ask'+col]].mean(axis=1)
conn.close()
data
```

```python
window = 14
data['mfi'] = ta.volume.money_flow_index(high=data['high'], low=data['low'], close=data['close'], 
                                         volume=data['tickqty'], window=window, fillna=False)
```

```python
data.dropna(inplace=True)
```

```python
data[['mfi', 'close']].iloc[-200:].plot(figsize=(15,5), secondary_y='mfi')
plt.show()
```

```python
data['signal'] = np.where(data['mfi'] > 80, -1,
                          np.where(data['mfi'] < 20, 1, np.nan))
data['signal'] = data['signal'].ffill().fillna(0)
```

```python
data['signal'].value_counts()
```

```python
data['signal'].diff(1).apply(np.abs).value_counts()
```

```python
backtest = IterativeBacktester(data, data['signal'], 'm5')
backtest.backtest()
backtest.results()
```
