```python
freq = 'm15'
conn = sqlite3.connect(f'../PriceData/PriceData_{freq}.db')
ticker = major_forex_pairs[4]
print(ticker)
data = pd.read_sql(f"SELECT * FROM '{ticker}'", conn, parse_dates=['date'], index_col=['date'])
conn.close()
columns = ['open', 'high', 'low', 'close']
for col in columns:
    data[col] = data[['bid'+col, 'ask'+col]].mean(axis=1)
display(data)
test_size = 0.048
test_len = int(test_size * len(data))
```

```python
period = (20, 100)
data['linreg1'] = talib.LINEARREG(data['close'], timeperiod=period[0])
data['linreg2'] = talib.LINEARREG(data['close'], timeperiod=period[1])
# data['slope'] = talib.LINEARREG_SLOPE(data['close'], timeperiod=period)
```

```python
# data['signal'] = data['slope'].apply(np.sign).mul(-1)
data['signal'] = np.where(data['linreg1'] > data['linreg2'], 1,
                          np.where(data['linreg2'] > data['linreg1'], -1, np.nan))
```

```python
if data.shape[0] > test_len:
    data = data.iloc[-test_len:]
else:
    pass
```

```python
data['signal'] = data['signal'].ffill().fillna(0)
```

```python
data[['close', 'linreg1', 'linreg2', 'signal']].plot(figsize=(15,5), secondary_y=['signal'])
```

```python
print('*** Positions ***')
print(data['signal'].value_counts())
print('-'*50)
print('*** Trades ***')
print(data['signal'].diff(1).apply(np.abs).value_counts())
```

```python
backtest = IterativeBacktester(data=data,
                               signals=data['signal'],
                               freq=freq)
backtest.backtest()
backtest.results()
```
