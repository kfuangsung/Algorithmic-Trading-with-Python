```python
import sys
sys.path.append('../')
import sqlite3
import pandas as pd 
import numpy as np 
import ta 
from backtester import *
```

```python
conn = sqlite3.connect('../PriceData/PriceData_m5.db')
data = pd.read_sql("SELECT * FROM 'EUR/USD'", conn, parse_dates=['date'], index_col=['date'])
for col in ['open', 'close', 'high', 'low']:
    data[col] = data[['bid'+col, 'ask'+col]].mean(axis=1)
data
```

```python
stochrsi = ta.momentum.StochRSIIndicator(close=data['close'], window=20, smooth1=5, smooth2=5, fillna=False)
data['stochrsi'] = stochrsi.stochrsi()
data['stochrsi_k'] = stochrsi.stochrsi_k()
data['stochrsi_d'] = stochrsi.stochrsi_d()
data.dropna(axis=0, inplace=True)
```

```python
data[['stochrsi', 'stochrsi_k', 'stochrsi_d']].iloc[-100:].plot(figsize=(15,5))
plt.axhline(y=0.2, color='k', linestyle='--')
plt.axhline(y=0.8, color='k', linestyle='--')
plt.show()
```

```python
data['signal'] = np.where(data['stochrsi'] > 0.8, -1,
                          np.where(data['stochrsi'] < 0.2, 1, np.nan))
data['signal'] = data['signal'].ffill().fillna(0)
```

```python
print('Positions')
print(data['signal'].value_counts())
print("-"*50)
print("Position changing")
print(data['signal'].diff(1).apply(np.abs).value_counts())
```

```python
backtest = IterativeBacktester(data, data['signal'], 'm5')
backtest.backtest()
backtest.results()
```
