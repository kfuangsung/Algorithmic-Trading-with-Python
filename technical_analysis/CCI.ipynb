```python
import sys
sys.path.append('../')
import sqlite3
import json
from backtester import *
from fxcmtoken import major_forex_pairs
from sklearn.model_selection import train_test_split
from datetime import datetime

major_forex_pairs
```

```python
freq = 'm15'
conn = sqlite3.connect(f'../PriceData/PriceData_{freq}.db')
ticker = major_forex_pairs[0]
print(ticker)
data = pd.read_sql(f"SELECT * FROM '{ticker}'", conn, parse_dates=['date'], index_col=['date'])
conn.close()
columns = ['open', 'high', 'low', 'close']
for col in columns:
    data[col] = data[['bid'+col, 'ask'+col]].mean(axis=1)
print(data.shape)
test_size = 0.047 # m15
test_len = int(test_size * len(data))
print("test length", test_len)
display(data)
```

```python
data['cci'] = ta.trend.cci(high=data['high'], low=data['low'], close=data['close'], window=50, fillna=False)
data.dropna(inplace=True)
```

```python
data['cci'].iloc[-400:].plot(figsize=(15,5))
```

```python
if data.shape[0] > test_len:
    data = data.iloc[-test_len:]
else:
    pass
```

```python
# strategy --> overbought/sold
# cci > 200 --> sell
# cci < -200 --> buy
# else NA
# then ffill
data['cci_signal'] = np.where(data['cci'] > 200, -1,
                     np.where(data['cci'] < -200, 1, np.nan))
data['cci_signal'] = data['cci_signal'].ffill().fillna(0)
```

```python
print(data['cci_signal'].value_counts())
print("-"*50)
print(data['cci_signal'].diff(1).apply(np.abs).value_counts())
```

```python
backtest = IterativeBacktester(data, data['cci_signal'], freq)
backtest.backtest()
backtest.results()
```
