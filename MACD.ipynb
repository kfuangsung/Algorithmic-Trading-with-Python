```python
import pandas as pd 
import numpy as np 
import ta
from backtester import *
```

```python
data = pd.read_csv('../forex_prices/m5/EUR_USD_m5.csv', index_col=['date'], parse_dates=['date'])
columns = ['open', 'high', 'low', 'close']
for col in columns:
    data[col] = data[['bid'+col, 'ask'+col]].mean(axis=1)
display(data.head())
print(data.shape)
```

```python
data['macd_diff'] = ta.trend.macd_diff(data['close'], window_slow=26, window_fast=12, window_sign=9, fillna=False)
data.dropna(inplace=True)
display(data.head())
print(data.shape)
```

```python
data['macd_diff'].iloc[-100:].plot(kind='bar', figsize=(15,5), xticks=[])
plt.show()
```

```python
data['macd_signal'] = np.where(data['macd_diff'] > 0, 1, np.where(data['macd_diff'] < 0, -1, 0))
print('Positions')
print(data['macd_signal'].value_counts())
print('-'*50)
print('Positions change')
print(data['macd_signal'].diff().apply(np.abs).value_counts())
print('-'*50)
```

```python
backtest = IterativeBacktester(data, data['macd_signal'], 'm5')
```

```python
backtest.backtest()
```
